# Análisis del Problema de Predicción de Casos de COVID-19 en Nueva York

## 1. ¿Cuál es el objetivo del problema?
El objetivo de este problema es analizar la evolución de los casos de COVID-19 en la ciudad de Nueva York, observando los **casos confirmados**, **hospitalizaciones**, y **muertes**. Se puede usar para predecir tendencias futuras, evaluar el impacto de intervenciones, o para informar a las autoridades de salud.

**Posibles objetivos específicos**:
- Predecir el número de hospitalizaciones o muertes futuras a partir de los casos confirmados.
- Identificar patrones de propagación del virus en diferentes distritos de la ciudad.
- Analizar el comportamiento de las tasas de crecimiento de casos y muertes.

## 2. ¿Cómo se usará su solución?
La solución puede ayudar a identificar **puntos críticos** en los distritos con mayor número de casos y optimizar los recursos médicos, como camas de hospital.

**Ejemplos de uso**:
- Alertas tempranas para hospitales.
- Proyectos de intervención en áreas con alta tasa de hospitalización o muertes.

## 3. ¿Cuáles son las soluciones actuales (si las hay)?
Las soluciones actuales incluyen **modelos de predicción basados en series temporales**, como ARIMA o modelos más complejos de **machine learning**. Además, las autoridades utilizan estadísticas diarias y modelos epidemiológicos básicos.

**Herramientas y soluciones actuales**:
- Modelos SEIR (Susceptible-Expuesto-Infectado-Recuperado).
- Modelos de regresión para predecir tendencias.
- Dashboards interactivos para monitorear casos en tiempo real.

## 4. ¿Cómo se debe enmarcar este problema?
- **Tipo de problema**: Supervisado (si se trata de una predicción de casos o muertes futuras).
- **En línea / Fuera de línea**: En este caso, podría ser **fuera de línea**, ya que se puede entrenar un modelo con los datos históricos y luego usarlo para hacer predicciones sobre datos futuros.
- **Técnicas a usar**:
  - **Regresión lineal o logística** para predecir muertes o hospitalizaciones.
  - **Redes neuronales o XGBoost** si el problema es más complejo.

## 5. ¿Cómo se debe medir el desempeño o el rendimiento de la solución, una primera intuición?
El desempeño se puede medir con **error absoluto medio (MAE)** o **error cuadrático medio (RMSE)**, que son métricas comunes para problemas de predicción.

**Métricas clave**:
- **RMSE**: Mide la diferencia promedio entre los valores predichos y los reales.
- **R^2**: Para evaluar qué tan bien se ajusta el modelo a los datos observados.

## 6. ¿La medida de desempeño está alineada con el objetivo del problema?
Sí, ya que tanto el MAE como el RMSE nos indican cuán cercanas son las predicciones a los valores reales, lo que es crucial para tomar decisiones informadas en salud pública.

## 7. ¿Cuál sería el desempeño o rendimiento mínimo necesario para alcanzar el objetivo del problema?
El rendimiento mínimo sería que el modelo sea lo suficientemente preciso como para predecir las tendencias diarias de casos y muertes con un margen de error aceptable, lo cual puede variar dependiendo de las necesidades de los responsables de la salud pública.

## 8. ¿Cuáles son los problemas parecidos? ¿Se puede reutilizar experiencias o herramientas ya creadas?
Problemas similares con **modelos epidemiológicos** como el SEIR, y herramientas de predicción de series temporales.

## 9. ¿Hay experiencia del problema disponible?
Sí, existen varios trabajos previos que han utilizado modelos para predecir la evolución de COVID-19 a partir de datos históricos, con herramientas como **Prophet**, **ARIMA**, y **redes neuronales recurrentes (RNN)**. Además, se han utilizado datos de diversas ciudades y países para estos modelos.

## 10. (Importante) ¿Cómo se puede resolver el problema manualmente?
Manual o descriptivamente, este problema podría abordarse mediante **sumarización de casos, hospitalizaciones y muertes**, y utilizando métodos estadísticos para calcular promedios de 7 días o tendencias. Sin embargo, las predicciones más complejas, como los pronósticos a futuro, requieren modelos computacionales.
