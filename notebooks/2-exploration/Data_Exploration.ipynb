```python
from pathlib import Path

import numpy as np
import pandas as pd
import pyarrow as pa
```

```python
BASE_DIR = Path().resolve().parents[1]  # sube dos niveles
DATA_DIR = BASE_DIR / "data" / "01_raw"
file_path = DATA_DIR / "covid_raw.csv"

# Leer el archivo CSV
coviddailiy_df = pd.read_csv(file_path)
```

```python
missing_values = ["", " ", "NA", "N/A"]
coviddailiy_df = coviddailiy_df.replace(missing_values, np.nan)
```

```python
MISSING_THRESHOLD = 0.8
missing_percentage = coviddailiy_df.isnull().sum() / len(coviddailiy_df)
columns_to_remove_missing = missing_percentage[
    missing_percentage > MISSING_THRESHOLD
].index.tolist()
print(" Columnas eliminadas por tener m谩s del 80% de valores faltantes:", columns_to_remove_missing)
coviddailiy_df = coviddailiy_df.drop(columns=columns_to_remove_missing)
```

```python
columns_to_remove_irrelevant = ["INCOMPLETE"]
print(" Columnas eliminadas por ser irrelevantes:", columns_to_remove_irrelevant)
for column in columns_to_remove_irrelevant:
    print(
        f"{column}: Esta columna no aporta informaci贸n relevante para el an谩lisis, "
        "esta informaci贸n fue tomada del diccionario de datos del dataset."
    )
coviddailiy_df = coviddailiy_df.drop(columns=columns_to_remove_irrelevant)
```
